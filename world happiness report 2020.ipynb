```python
subset = df[['Country name','Social support','Healthy life expectancy', 'Freedom to make life choices', 'Generosity', 'Perceptions of corruption']]
```

```python
subset.head()
```

```python
subset = subset.set_index('Country name')
```

```python
subset.head()
```

```python
import seaborn as sns
import matplotlib.pyplot as plt
%matplotlib inline 
fig, ax = plt.subplots(figsize=(20,20))
sns.heatmap(subset.corr(),ax=ax,annot=True,linewidth=0.04, fmt='.2f',cmap='magma')
plt.show()
```
