```python
import seaborn as sns
import matplotlib.pyplot as plt
%matplotlib inline 
fig, ax = plt.subplots(figsize=(10,10))
sns.heatmap(happiness_report.corr(),ax=ax,annot=True,linewidth=0.05,fmt='.2f',cmap='magma')
plt.show()
```
